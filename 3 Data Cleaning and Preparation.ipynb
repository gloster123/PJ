```python

df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df['gender'].unique())
```

```python

# Ensure TotalCharges is treated as a numerical value
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
print("Original dataset shape:", df.shape)
```

```python
plt.boxplot(df['TotalCharges'] )
plt.show()
```

```python

Q1 = df['TotalCharges'].quantile(0.25)
Q3 = df['TotalCharges'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['TotalCharges'] < (Q1 - 1.5 * IQR)) | (df['TotalCharges'] > (Q3 + 1.5 * IQR))]
df = df[~df.isin(outliers)]

# Display the shape of the dataset before and after removing outliers

print("Cleaned dataset shape:", df.shape)
```

```python
# feature enginering 
# Calculate average monthly charges
df['AvgMonthlyCharges'] = df['TotalCharges'] / df['tenure']
df
```

```python
from sklearn.preprocessing import StandardScaler

# Scale numerical features
scaler = StandardScaler()
df[['MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(df[['MonthlyCharges', 'TotalCharges']])
```

```python
from sklearn.model_selection import train_test_split

# Define features and target
x = df.drop('Churn', axis=1)
y = df['Churn']

# Split the data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Export the cleaned training and testing sets
xtrain.to_csv("Cleaned_Telecom_Customer_Churn_Train.csv", index=False)
xtest.to_csv("Cleaned_Telecom_Customer_Churn_Test.csv", index=False)
ytrain.to_csv("Cleaned_Telecom_Customer_Churn_y_Train.csv", index=False)
ytest.to_csv("Cleaned_Telecom_Customer_Churn_y_Test.csv", index=False)
```
