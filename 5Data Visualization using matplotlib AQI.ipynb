```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("city_day.csv")
df
```

```python
df.isnull().sum()
```

```python
# Impute missing values for pollutants and AQI using forward fill
df['PM2.5'].fillna(method='ffill', inplace=True)
df['PM10'].fillna(method='ffill', inplace=True)
df['CO'].fillna(method='ffill', inplace=True)
df['AQI'].fillna(method='ffill', inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# For categorical columns like 'AQI_Bucket', fill with mode
df['AQI_Bucket'].fillna(df['AQI_Bucket'].mode()[0], inplace=True)
```

```python
# Check remaining missing values
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 2))
plt.plot(df['Date'], df['AQI'], color='blue')
plt.title('Overall AQI Trend Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
# plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,2))
plt.plot(df['Date'], df['PM2.5'] )
plt.title(f' Levels Over Time')
plt.xlabel('Date')
plt.ylabel(f' Levels')
```

```python
plt.figure(figsize=(12,2))
plt.plot(df['Date'], df['PM10'] )
plt.title(f' Levels Over Time')
plt.xlabel('Date')
plt.ylabel(f' Levels')
```

```python
plt.figure(figsize=(12,2))
plt.plot(df['Date'], df['CO'] )
plt.title(f' Levels Over Time')
plt.xlabel('Date')
plt.ylabel(f' Levels')
```

```python
# Set the time frame if necessary, e.g., last 30 days
recent_data = df.tail(30)

plt.figure(figsize=(12, 3))
plt.bar(recent_data['Date'], recent_data['AQI'], color='skyblue')
plt.title('AQI Comparison Over Recent Dates')
plt.xlabel('Date')
plt.ylabel('AQI')
# plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(data=df[['PM2.5', 'PM10', 'CO','AQI']])
plt.title('Distribution of Pollutant Levels')
plt.ylabel('Values')
plt.xlabel("Pollutants")
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Scatter plot for AQI and PM2.5
plt.subplot(1, 3, 1)
plt.scatter(df['PM2.5'], df['AQI'], color='purple')
plt.title('AQI vs PM2.5')
plt.xlabel('PM2.5')
plt.ylabel('AQI')

# Scatter plot for AQI and PM10
plt.subplot(1, 3, 2)
plt.scatter(df['PM10'], df['AQI'], color='green')
plt.title('AQI vs PM10')
plt.xlabel('PM10')
plt.ylabel('AQI')

# Scatter plot for AQI and CO
plt.subplot(1, 3, 3)
plt.scatter(df['CO'], df['AQI'], color='orange')
plt.title('AQI vs CO')
plt.xlabel('CO')
plt.ylabel('AQI')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 3))
plt.plot(df['Date'], df['AQI'], color='darkred', linestyle='--', marker='o', label='AQI')
plt.title('Customized AQI Trend')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
