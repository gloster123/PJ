```python
# Convert ORDERDATE to datetime if itâ€™s not already
df1['ORDERDATE'] = pd.to_datetime(df1['ORDERDATE'])
```

```python
# Deriving TOTALSALES as QUANTITYORDERED * PRICEEACH
df1['TOTALSALES'] = df1['QUANTITYORDERED'] * df1['PRICEEACH']
```

```python
total_sales = df1['TOTALSALES'].sum()
avg_order_value = df1['TOTALSALES'].mean()
sales_by_category = df1.groupby('PRODUCTLINE')['TOTALSALES'].sum()
```

```python
import matplotlib.pyplot as plt

# Total sales by product line
sales_by_category.plot(kind='bar', color='skyblue', title='Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Distribution of deal sizes
df1['DEALSIZE'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Deal Size Distribution')
plt.show()
```

```python
df = df1
df.groupby(df['ORDERDATE'].dt.to_period('M'))['TOTALSALES'].sum().plot(kind='line', title='Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
